```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

**TASK: Plot the color histograms for the RED, BLUE, and GREEN channel of the giaraffe image. Pay careful attention to the ordering of the channels.**

```python
# read image
picture =cv2.imread('DATA/giraffes.jpg')

# create a loop for r g b iteration
color = ('b','g','r')
for i,col in enumerate(color):
    histogram = cv2.calcHist([picture],[i],None,[256],[0,256])
    plt.plot(histogram,color=col)
    
# display the result
plt.title("Histogram Plotted Of Girrafe")
plt.show()
```

```python
#OPEN PICTURE AND CONVERT TO RGB
picture1 = cv2.imread('DATA/giraffes.jpg')
picture1 = cv2.cvtColor(picture1,cv2.COLOR_BGR2RGB)

picture2 = cv2.imread('DATA/copyright_gautam.png')
picture2 = cv2.cvtColor(picture2,cv2.COLOR_BGR2RGB)
```

```python
# Resize Both pic as same resolution

picture1 = cv2.resize(picture1,(500,400))
picture2 = cv2.resize(picture2,(500,400))
```

```python
# blend pic2 over pic 1
blend_pic = cv2.addWeighted(src1=picture1,alpha=0.8,src2=picture2,beta=0.2,gamma=30)

#display
display_img(blend_pic)
```

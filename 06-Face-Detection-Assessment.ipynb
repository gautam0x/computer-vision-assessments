# Face Detection 

```python
# import useful libraries
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# define face_haarcascade for face detection
face_cascade = cv2.CascadeClassifier("DATA/haarcascades/haarcascade_frontalface_default.xml")
```

```python
# open image in grayscale mode
person_image = cv2.imread("DATA/tony_dp.jpg",0)

# get the point to draw rectangle on face
face_rect = face_cascade.detectMultiScale(person_image)

# draw rectangle on face
for (x,y,w,h) in face_rect:
    cv2.rectangle(person_image,(x,y), (x+w,y+h),(255,255,255),10)
```

```python
# Display The Rusult
plt.imshow(person_image,cmap='gray')
```

# Group Face Detection With Custom Parameter

```python
# define face_haarcascade for face detection
face_cascade = cv2.CascadeClassifier("DATA/haarcascades/haarcascade_frontalface_default.xml")
```

```python
#Open Group Of Peoples Picture in RGB format
group_img = cv2.imread('DATA/baahubali_cast.jpg') # in BGR
group_img = cv2.cvtColor(group_img,cv2.COLOR_BGR2RGB)
```

```python
# Define Custom Parameter 
scale_factor = 1.3
min_neighbour = 5

# get the point to draw rectangle on eye
face_rect = face_cascade.detectMultiScale(group_img,scale_factor,min_neighbour)

# draw rectangle on face
for (x,y,w,h) in face_rect:
    cv2.rectangle(group_img,(x,y), (x+w,y+h),(255,0,0),10)
```

```python
# Display Output
plt.imshow(group_img)
```

# Face Detection In Live Video

**Capture And Display**

```python
import cv2

# Define Function for detection
def detect_face(img):
    # Defince Cascade File
    face_cascade = cv2.CascadeClassifier("DATA/haarcascades/haarcascade_frontalface_default.xml")
    
    # Draw Rectangle on image
    face_rect = face_cascade.detectMultiScale(img,scaleFactor=1.2,minNeighbors=4)
    for (x,y,w,h) in face_rect:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),10)
        
    return img


# Start Program
capture = cv2.VideoCapture(0)

while True:
    
    ret,frame = capture.read(0)
    output_img = detect_face(frame)
    
    cv2.imshow("Window",output_img)
    
    c = cv2.waitKey(1)
    if c == 27:
        break

capture.release()
cv2.destroyAllWindows()
```
